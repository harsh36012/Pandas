```python
# date_range function
```

```python
# generate daily dates in a given range
pd.date_range(start='2026-1-23',end='2026-2-15',freq='D')  # freq = D ,each day
```

```python
pd.date_range(start='2026-1-23',end='2026-2-15',freq='2d')  #freq = 2D, every 2nd day
```

```python
pd.date_range(start = '2025, 5,1' , end = '2025, 7,29' ,freq = '3D')
```

```python
# B -> business days
pd.date_range(start='2023/1/5',end='2023/2/28',freq='B')  # only business day (mon-fri)
```

```python
# W -> one week per day
pd.date_range(start='2023/1/5',end='2023/2/28',freq='W')  # only 1 day a week (by default W-SUN {sunday}) 
```

```python
pd.date_range(start='2023/1/5',end='2023/2/28',freq='W-WED')  #  only wednesday 
```

```python
# H -> Hourly data(factor)
pd.date_range(start='2023/1/5',end='2023/2/28',freq='h')  # every 1 hour
```

```python
pd.date_range(start='2023/1/5',end='2023/2/28',freq='6h')  # every 6 hour
```

```python
# ME -> Month end
pd.date_range(start='2023/1/5',end='2023/4/28',freq='ME')  # every month end
```

```python
# MS -> Month start
pd.date_range(start='2023/1/5',end='2023/3/28',freq='MS')  # each month 1st day 
```

```python
# YE -> Year end
pd.date_range(start='2023/1/5',end='2030/2/28',freq='YE')  # every year
```

```python
# using periods(number of results)
pd.date_range(start='2023/1/5',periods=25,freq='ME')
```

```python
# to_datetime()
# convert object to Timestamp/DatetimeIndex
```

```python
## simple series example

s = pd.Series(['2023/1/1','2022/1/1','2021/1/1'])
s
```

```python
pd.to_datetime(s)
```

```python
pd.to_datetime(s).dt.day_name()
```

```python
# with errors
s = pd.Series(['2023/1/1','2022/1/1','2021/130/1'])
s
```

```python
# if we convert it to timestamp
pd.to_datetime(s)  # error  bcz of -> 2021/130/1
```

```python
# so to handle it by  errors='coerce'
pd.to_datetime(s,errors='coerce')
```

```python
pd.to_datetime(s,errors='coerce').dt.month_name()
```

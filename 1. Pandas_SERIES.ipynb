```python
import numpy as np
import pandas as pd
```

## Creating Series

### Series from List

```python
# String Type
country = ["India","USA","Japan","Nepal","China"]

country_series = pd.Series(country)
country_series
```

```python
# Integer Type
# Custom index
marks = [99,85,89,72,71]
subjects = ["Computer Science","Physics","Chemistry","English","Maths"]
pd.Series(marks,index=subjects)
```

```python
# Setting a name
Marks_series = pd.Series(marks,index = subjects,name = "Mine 12th Marks")
Marks_series
```

### Series from Dict

```python
# marks_series = {
#     "Computer Science" : 99,
#     "Physics" : 85,
#     "Chemistry" : 89,
#     "English" : 72,
#     "Maths" : 71
# }
# Marks_series = pd.Series(marks_series, name= "Harsh's 12th Marks")
# Marks_series

marks = {"cs":99,"maths": 71,"phy":85,"chem":89,"eng":72}
marks_series = pd.Series(marks,name= "12th Marks")
marks_series
```

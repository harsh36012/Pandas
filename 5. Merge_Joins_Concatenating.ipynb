```python
# right join

temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
students.merge(regs,how='right',on='student_id')  # all values of students will come regardless of join
```

```python
regs.merge(students,how='left',on='student_id')
```

```python
# outer join / full join
students.merge(regs,how='outer',on='student_id')  # all the elements of both the tables will come regardless of join
```

```python
# find top 3 studiums with highest sixes/match ratio
temp_df = delivery.merge(matches,left_on="match_id",right_on="id")
temp_df
```

```python
six_df = temp_df[temp_df["batsman_runs"] == 6]
six_df
```

```python
num_six = six_df.groupby("venue")["venue"].count()
num_six
```

```python
num_matches = matches["venue"].value_counts()
num_matches
```

```python
(num_six/num_matches).sort_values(ascending=False).head(3)
```

```python
# find orange cap holder of all the seasons
temp_df.groupby(["season",'batsman'])['batsman_runs'].sum()
```

```python
temp_df.groupby(["season",'batsman'])['batsman_runs'].sum().reset_index()
```

```python
temp_df.groupby(["season",'batsman'])['batsman_runs'].sum().reset_index().sort_values("batsman_runs",ascending=False)
```

```python
temp_df.groupby(["season",'batsman'])['batsman_runs'].sum().reset_index().sort_values("batsman_runs",ascending=False).drop_duplicates(subset=["season"]).sort_values("season")
```

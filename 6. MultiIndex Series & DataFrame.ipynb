```python
index_val = [("cse",2019),("cse",2020),("cse",2021),("cse",2021),("ece",2019),("ece",2020),("ece",2021),("ece",2021)]
a = pd.Series([1,2,4,1,2,4,3,2],index=index_val)  
a

# this is not a correct way to create.
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# Create multiIndex object

# using pd.MultiIndex.from_tuples()
index_val = [("cse",2019),("cse",2020),("cse",2021),("cse",2022),("ece",2019),("ece",2020),("ece",2021),("ece",2021)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
multiindex.levels[1]
```

```python
multiindex.levels[0]
```

```python
# another way
# using -> pd.MultiIndex.from_product()

multiindex = pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
multiindex
```

```python
# creating a series using multiindex object
b = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
b
```

```python
b[('cse',2019)]
```

```python
b['cse']
```

```python
# unstack -> under vala index ko lekr col me dal deta h, if pahle se h then, 1 level or under krega like this.
print(b)
temp = b.unstack()
temp
```

```python
# stack -> under vala col ko leta h.or row me dal deta h, if pahle se h to 1 level or under dal dega
temp.stack()
```

```python
# melt -> real world example
death = pd.read_csv('time_series_covid19_deaths_global.csv')
confirm = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
death
```

```python
confirm
```

```python
# we want cols -> country,date, death, confirm
```

```python
death = death.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='date',value_name='num_deaths')
death
```

```python
confirm = confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='date',value_name='num_confirm')
confirm
```

```python

death.merge(confirm,on=['Province/State','Country/Region','Lat','Long','date'])[['Country/Region','date','num_deaths','num_confirm']]
```

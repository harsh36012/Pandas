```python
# split , get

df['Name']

# in this Name :-  lastname, tag. lastname -> this is how all names are stored.
# we want separate cols for firstName , Tag, LastName
```

```python
df['LastName'] = df['Name'].str.split(",").str.get(0)
```

```python
df
```

```python
df[['title','FirstName']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
```

```python
df
```

```python
df['title'].value_counts()
```

```python
# replace 
# Miss. / Ms. / Mile -> all are same.
# so we want replace Ms./ Mlle with Miss.

df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith

# get the FirstName whose starting letter is A
df[df['FirstName'].str.startswith('A')]
```

```python
# get the name whose ending letter is A
df[df['FirstName'].str.endswith('A')]
```

```python
# get whose Name contains only digit
df[df['Name'].str.isdigit()]
```

```python
# get whose name contains only aplhabet 
df[df['Name'].str.isalpha()]
```

```python
# get the names whose name contains john(in any case)
df[df['FirstName'].str.contains("john",case=False)]
```

```python
# get the names whose lastName ->  starts with vowel and ends with vowel
# using regural expression

df[df['LastName'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
# last name who starts with consonent and ends with consonent
df[df['LastName'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
